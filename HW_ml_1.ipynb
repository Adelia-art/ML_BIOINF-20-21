# Загрузка данных

```python
def read_covid_dataset(path_to_csv):
    data_open = pd.read_csv(path_to_csv)
    data_open['target']=['None' for x in data_open ['date']]
    data = data_open.loc [1:10000]
    X=data[data.columns.difference(['target'])].values
    y = data.target.values   
    return X, y

def read_hr_dataset(path_to_csv):
    data_open = pd.read_csv(path_to_csv)
    data = data_open.loc [1:10000] 
    X=data[data.columns.difference(['target'])].values
    y = data.target.values  
    return X, y
```

```python
COVID_DATA_PATH = "covid.csv"
```

```python
HR_DATA_PATH = "hr.csv"
```

```python
X_covid, _ = read_covid_dataset(COVID_DATA_PATH)
X_hr, y_hr = read_hr_dataset(HR_DATA_PATH)
```
